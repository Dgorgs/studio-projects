```python
#create a variable to convert the string into a datetime
day_1 = datetime.datetime.strptime('06/29/2015', "%m/%d/%Y")
day_2 = datetime.datetime.strptime('06/30/2015', "%m/%d/%Y")
#requests[(requests['created_at']>day_1) & (requests['created_at']<day_2)].plot(kind="bar")
```

```python
#requests[(requests['created_at']>day_1) & (requests['created_at']<day_2)].plot() #(y='Complaint Type', x='Created Date')
```

```python
twelve_hours = requests[(requests['created_at']>day_1) & (requests['created_at']<day_2)]
twelve_hours['Complaint Type'].head()
```

```python
twelve_hours.set_index('created_at', inplace=True)
```

```python
twelve_hours.head()
```

```python
#soma's command
twelve_hours.groupby(by=[twelve_hours.index.hour, 'Complaint Type']).size().reset_index().pivot(index='created_at', columns='Complaint Type', values=0).fillna(0).plot(stacked=True, kind="area", figsize=(30,10))
plt.savefig("twelve_hours.pdf")
```
